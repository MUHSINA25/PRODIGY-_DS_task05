```python
# Convert 'Start_Time' to datetime format
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Extract hour, day, month, and year from 'Start_Time'
df['Hour'] = df['Start_Time'].dt.hour
df['Day'] = df['Start_Time'].dt.day
df['Month'] = df['Start_Time'].dt.month
df['Year'] = df['Start_Time'].dt.year
```

## EDA

```python
# Severity Distribution
sns.countplot(x='Severity', data=df, palette='Set2')
plt.title('Severity Distribution')
plt.show()
```

###### This bar chart displays the distribution of severity levels, showing a stark contrast between two categories (severity levels 2 and 4). Level 2 severity dominates the distribution with approximately 95,000 cases, while severity level 4 has a much smaller count of around 1,000-2,000 cases. This highly skewed distribution suggests that whatever is being measured, moderate severity (level 2) incidents or cases are far more common than high severity (level 4) ones. The absence of levels 1 and 3 could either indicate that these levels don't exist in the classification system, or no cases were recorded for these levels. This type of distribution is often seen in incident reporting systems where serious cases (level 4) are relatively rare compared to more moderate cases (level 2).

```python
# Histogram for numeric features
df[['Temperature(F)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Distance(mi)']].hist(figsize=(12, 8), bins=30, color='skyblue', edgecolor='black')
plt.suptitle('Distribution of Numeric Features')
plt.show()
```

###### This set of four histograms shows the distribution of different weather-related features, each revealing distinct patterns:
Temperature(F): Shows a roughly normal distribution centered around 50-60Â°F, with temperatures ranging from about -20Â°F to 90Â°F. This suggests a dataset that spans different seasons or captures temperature variations across a wide geographic area.
Visibility(mi): Has a dramatic spike around the 10-mile mark, with much lower frequencies at other distances. This could indicate that clear visibility conditions (around 10 miles) are most common in the dataset, with fewer instances of poor visibility.
Wind_Speed(mph): Displays a right-skewed distribution, with most wind speeds concentrated between 0-15 mph. There's a sharp peak around 5-10 mph, and the frequency decreases rapidly as wind speed increases. This is typical of wind speed patterns where moderate winds are common and high winds are rare.
Distance(mi): Shows an extremely skewed distribution with the vast majority of measurements falling in the 0-20 mile range. This suggests that most events or measurements in the dataset occur within a short distance, possibly indicating local weather monitoring or reporting patterns.
Together, these distributions paint a picture of what appears to be a weather or environmental monitoring dataset, with most conditions being moderate and extreme conditions being relatively rare.

```python
# Categorical Feature Analysis
categorical_cols = ['Weather_Condition', 'State', 'Sunrise_Sunset']
for col in categorical_cols:
    plt.figure(figsize=(10, 5))
    df[col].value_counts().head(10).plot(kind='bar', color='orange', edgecolor='black')
    plt.title(f'Distribution of {col}')
    plt.show()
```

###### These three bar charts reveal interesting patterns about weather conditions, state distribution, and day/night occurrences:

###### Weather Condition Distribution: Shows that clear weather ("F") is overwhelmingly the most common condition, followed by cloudy weather. There's a declining frequency pattern through various other weather conditions like thunderstorms, light rain, and snow. This suggests that severe or adverse weather conditions are relatively rare in this dataset.
###### State Distribution: Displays what appears to be a geographic distribution across different US states, with two states (likely represented by 'CA' and 'FL' based on the pattern) having significantly higher frequencies than others. This could indicate either population density differences or varying reporting frequencies across states.

###### Sunrise/Sunset Distribution: Shows roughly a 60/40 split between "Day" and "Night" occurrences. This is interesting as it doesn't follow an exact 50/50 split that might be expected, suggesting there could be seasonal variations or reporting bias in when events are recorded.

###### Together, these distributions paint a picture of a dataset that's heavily weighted toward fair weather conditions, shows geographic concentration in a few states, and has a slight bias toward daytime occurrences. This could be useful for understanding patterns in weather-related incidents or events.

```python
# Severity vs Weather Conditions
plt.figure(figsize=(15, 6))
sns.countplot(x='Weather_Condition', hue='Severity', data=df, order=df['Weather_Condition'].value_counts().head(10).index)
plt.title('Severity vs Weather Conditions')
plt.xticks(rotation=45)
plt.show()
```

###### This bar chart compares the relationship between weather conditions and incident severity (levels 2 and 4), revealing several key insights:

###### Fair weather dominates the dataset with the highest number of incidents (around 35,000), primarily with severity level 2 (lighter color). This is followed by cloudy and mostly cloudy conditions, showing a clear decreasing trend in incident frequency as weather conditions become more adverse.

###### An important pattern emerges: across all weather conditions, severity level 2 (lighter shade) incidents vastly outnumber severity level 4 (darker shade) incidents. This remains consistent from fair weather through to haze, suggesting that severe (level 4) incidents are relatively rare regardless of weather conditions.

###### The frequency of incidents decreases notably as weather conditions become more severe or challenging (moving from fair to heavy rain and haze). However, the proportion of severity 4 incidents doesn't appear to increase significantly in adverse weather, which might be counterintuitive.

###### This visualization suggests that while weather conditions influence the overall frequency of incidents, they don't necessarily correlate with increased severity. Fair and cloudy conditions see the most incidents likely because these are the most common weather conditions, rather than these conditions being inherently more dangerous.

```python
# Severity by State
plt.figure(figsize=(15, 6))
sns.countplot(x='State', hue='Severity', data=df, order=df['State'].value_counts().index)
plt.title('Severity by State')
plt.xticks(rotation=90)
plt.show()
```

###### This visualization reveals a striking pattern in incident distribution across US states, with two states (likely California and Florida) significantly leading in incident counts. Both states show predominantly severity level 2 incidents (light pink), with California recording around 25,000 cases and Florida following with approximately 17,000. The remaining states show a dramatic drop-off in numbers, forming a long tail pattern with most states reporting under 5,000 incidents. Notably, severity level 4 incidents (dark color) remain consistently low across all states, suggesting that severe incidents are rare regardless of location. This pattern could reflect differences in population density, reporting systems, or traffic volumes rather than inherent safety differences between states.

```python
# Convert Start_Time to datetime
df['Start_Time'] = pd.to_datetime(df['Start_Time'])

# Extract Hour
df['Hour'] = df['Start_Time'].dt.hour

# Plot
plt.figure(figsize=(10, 5))
sns.countplot(x='Hour', hue='Severity', data=df, palette='viridis')
plt.title('Severity vs Hour of Day')
plt.xlabel('Hour of Day')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Start_Lng', y='Start_Lat', hue='Severity', data=df)
plt.title('Geographical Distribution of Severity')
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title="Severity")
plt.show()
```

###### This scatter plot shows incidents across the United States, using latitude and longitude. Incidents are marked by severity levels 2 (light pink) and 4 (dark blue).

###### Key patterns:

###### Most incidents are along the coasts, especially in California and Florida.
###### The Eastern seaboard has a dense cluster of incidents from Florida to the Northeast.
###### Urban areas have more incidents, shown as dense clusters.
###### Severity level 2 incidents dominate, while severity level 4 is scattered across the country.
###### Gaps are seen in less populated central areas, suggesting incident reporting is linked to population density and transport routes.
###### Overall, incidents are more common in urban and coastal areas, with no clear pattern for severe incidents.
